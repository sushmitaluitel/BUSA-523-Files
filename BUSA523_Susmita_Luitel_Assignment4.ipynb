1. Using the dataset provided, develop the following visualizations using the right set of variables.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#import the data
data = pd.read_csv("data.csv", header = 0)
print(data)
```

a. Histogram

```python
#Histogram for Fuel Price
plt.hist(data["Fuel_Price"])
plt.title("Histogram of Fuel Price")
plt.xlabel("Fuel Price")
plt.show()
```

b. Bar chart

```python
data.plot(x="Date", y = ['dept_1 sales', 'dept_2 sales', 'dept_3 sales'], kind='bar')
```

c. Linechart

```python
data.plot(x="Date", y = ['Temperature', 'Fuel_Price'])
```

d. Area chart

```python
data.plot.area(x="Date", y = 'Unemployment')
```

e. Scatter plot

```python
data.plot.scatter(x = 'Temperature',y = 'Fuel_Price')
```
